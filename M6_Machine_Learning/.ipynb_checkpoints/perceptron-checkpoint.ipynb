## Perceptrón


Clasificar  estudiantes: con necesidades especiales(2) regulares(1) utilizando caracteristicas como: edad, grado y calificacion. Esto permitira definir si un alumno debe recibir atencion especial



```python
#Manejo de datos
import pandas as pd

#Manejo de matrices
import numpy as np

#Graficas
import matplotlib.pyplot as plt

#Algoritmo
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("estudiantes_sup.csv")
df
```

```python
#Agrupacion de datos
x=df.iloc[:,0:3].values
y=df.iloc[:,3].values
print(x)
y
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
for i in range(0,len(y)):
    if y[i]==1:
        c='blue'
    if y[i]==2:
        c='red'
    ax.scatter(x[i,0],x[i,1],x[i,2],color=c)
#ax.scatter(xc[:,0],xc[:,1],xc[:,2],color="green")
ax.set_xlabel('Edad')
ax.set_ylabel('Grado')
ax.set_xlabel('Calificacion')
plt.show()
```

```python
#Implementacion del perceptron
pcp=Perceptron()
pcp.fit(x,y)
yp=pcp.predict(x)
df['modeloperc']=pd.Series(yp)
df
```

```python
print('Perceptron')
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
for i in range(0,len(yp)):
    if yp[i]==1:
        c='blue'
    if yp[i]==2:
        c='red'
    ax.scatter(x[i,0],x[i,1],x[i,2],color=c)
#ax.scatter(xc[:,0],xc[:,1],xc[:,2],color="green")
ax.set_xlabel('Edad')
ax.set_ylabel('Grado')
ax.set_xlabel('Calificacion')
plt.show()
```

```python
#Implementacion de la red neuronal
std=StandardScaler()
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4,random_state=0)
xtrain=std.fit_transform(xtrain)
xtest=std.fit_transform(xtest)
xu=std.fit_transform(x)
NNA=MLPClassifier(random_state=1,max_iter=300).fit(xtrain,ytrain)
ypnn=NNA.predict(xu)
df['modelonn']=pd.Series(ypnn)
df
```

```python
print('Redes neuronales')
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
for i in range(0,len(yp)):
    if ypnn[i]==1:
        c='blue'
    if ypnn[i]==2:
        c='red'
    ax.scatter(x[i,0],x[i,1],x[i,2],color=c)
#ax.scatter(xc[:,0],xc[:,1],xc[:,2],color="green")
ax.set_xlabel('Edad')
ax.set_ylabel('Grado')
ax.set_xlabel('Calificacion')
plt.show()
```
