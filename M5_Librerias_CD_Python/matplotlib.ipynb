```python
import sys
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
import numpy as np

print("Cargado")
```

```python
%matplotlib inline           
#activar modo interactivo de Matplotlib añadiendo la siguiente línea antes de generar tu gráfica

x=np.array(["comedia","accion","terror"])
y=np.array([47,23,86])
colores=["blue","red","mediumaquamarine"]

plt.bar(x,y,color=colores)        #Grafico de barras
plt.show()                        #Muestra la grafica
plt.savefig('grafic.png')        #Guardar la grafica
#sys.stdout.flush() 
```

```python
x1=np.random.normal(20,10,100)         #valores aleatorios, 100 nums, media de 20 y desviacion de 10
x2=np.random.normal(20,10,100)
x=[x1,x2]
#print(x1)
#print(x2)
#print(x1.shape)

#histograma
fig,ax=plt.subplots()                        # Crear figura(contiene la figura completa) y ejes(trazar la gráfica de línea)
ax.hist(x,color=["darkmagenta","mediumturquoise"])
plt.show()

```

```python
np.random.seed(5)

x1=np.random.normal(0,1,200)            #0 media, 1 desviacion, 200 numeros aleatorios
x2=np.random.normal(1,1,250)
x3=np.random.normal(5,1,100)
x=[x1,x2,x3]
fig,ax=plt.subplots()
%matplotlib inline           
plt.boxplot(x,whiskerprops=dict(color="blue", linewidth=2))
plt.show()
```

```python
x3=([1,1,2,3,8,7])
y3=([2,6,5,4,1,5])
plt.scatter(x3,y3, c="cyan",label="G1")

x4=([3,2,3,5,7,9])
y4=([2,4,6,5,3,1])
plt.scatter(x4,y4, c="purple",label="G2")    #Grafica de dispersion, color y etiqueta 

%matplotlib inline
plt.legend()
plt.show()
```

```python
X,Y=np.meshgrid(np.linspace(0,12,180),np.linspace(0,20,100))     #180 nums de 0 a 12, crear superficie/grilla
Z=np.sqrt(X**2+Y**2)

fig,ax=plt.subplots()
cnt=ax.contourf(X,Y,Z)                  #grafico de contorno relleno

cbar=ax.figure.colorbar(cnt,ax=ax)
cbar.ax.set_ylabel("Temperatura",rotation=-90,va="bottom")
plt.show()
```

```python
np.random.seed(50)
data=np.random.random((8,8))    #nums aleatorios de 8x8
#print(data)
fig,ax=plt.subplots()
im=ax.imshow(data)

cbar=ax.figure.colorbar(im,ax=ax)
cbar.ax.set_xlabel("semana",rotation=0, va="top")
cbar.ax.set_ylabel("produccion %",rotation=-90, va="bottom")
plt.show
```
